```python
#week 1
import pulp

# Create a linear programming problem
lp_problem = pulp.LpProblem("My_LP_Problem", pulp.LpMinimize)

# Define decision variables
units = [pulp.LpVariable(f"units{i}", lowBound=0) for i in range(1, 11)]
ot = [pulp.LpVariable(f"ot{i}", lowBound=0) for i in range(1, 11)]
z = [pulp.LpVariable(f"z{i}", cat=pulp.LpBinary) for i in range(1, 11)]

# Objective function
lp_problem += pulp.lpSum(ot), "Objective_Function"

print("---WEEK 1---")

# Constraints
lp_problem += units[0] + units[1] >= 3500
lp_problem += units[2] + units[3] + units[4] >= 3000
lp_problem += units[5] >= 4000
lp_problem += units[6] + units[7] >= 4000
lp_problem += units[8] + units[9] >= 2800

for i in range(10):
    lp_problem += units[i] <= z[i] * [3500, 3500, 3000, 3000, 4000, 4000, 4000, 4000, 2800, 2800][i]
    lp_problem += ot[i] <= z[i] * 48

# Additional Constraints
lp_problem += 1/40 * 1/0.6 * units[0] + z[0] * 8 + 1/60 * 1/0.65 * units[6] + z[6] * 8 <= 120 + ot[0] + ot[6]
lp_problem += 1/35 * 1/0.6 * units[1] + z[1] * 10 + 1/25 * 1/0.55 * units[2] + z[2] * 8 <= 120 + ot[1] + ot[2]
lp_problem += 1/30 * 1/0.55 * units[3] + z[3] * 10 + 1/45 * 1/0.60 * units[8] + z[8] * 24 <= 120 + ot[3] + ot[8]
lp_problem += 1/35 * 1/0.55 * units[4] + z[4] * 8 + 1/50 * 1/0.75 * units[5] + z[5] * 12 <= 120 + ot[4] + ot[5]
lp_problem += 1/60 * 1/0.65 * units[7] + z[7] * 8 + 1/50 * 1/0.60 * units[9] + z[9] * 20 <= 120 + ot[7] + ot[9]

lp_problem += ot[0] + ot[6] <= 48
lp_problem += ot[1] + ot[2] <= 48
lp_problem += ot[3] + ot[8] <= 48
lp_problem += ot[4] + ot[5] <= 48
lp_problem += ot[7] + ot[9] <= 48

# Solve the linear programming problem
lp_problem.solve()

# Print the results
print("Status:", pulp.LpStatus[lp_problem.status])
print("Optimal values:")
for i in range(10):
    print(f"units{i+1} = {pulp.value(units[i])}")

for i in range(10):
    print(f"ot{i+1} = {pulp.value(ot[i])}")

for i in range(10):
    print(f"z{i+1} = {pulp.value(z[i])}")

print("Optimal objective value:", pulp.value(lp_problem.objective))
```

```python
import pulp

initial_demand = [3500, 3000, 4000, 4000, 2800]

second_week_demand = [3500, 3000, 4000, 4000, 2800]

# Machine capacity for two weeks
machine_capacity = [120, 120, 120, 120, 120]

production_rates = {
    (1, 1): 40, (1, 2): 60,
    (2, 1): 35, (2, 2): 25,
    (3, 3): 45,
    (4, 2): 35, (4, 3): 50,
    (5, 4): 60, (5, 5): 50
}
yield_factors = {1: 0.6, 2: 0.55, 3: 0.75, 4: 0.65, 5: 0.6}

initial_setup = {
    1: 1,
    2: 2,
    3: 5,
    4: 3,
    5: 4
}

# Define the problem as a linear programming problem
prob_two_weeks = pulp.LpProblem("Production_Schedule_Two_Weeks", pulp.LpMinimize)

# Define decision variables for two weeks
x_two_weeks = pulp.LpVariable.dicts("Production", ((i, j, w) for i in range(1, 6) for j in range(1, 6) for w in range(2)), cat="Integer", lowBound=0)
y_two_weeks = pulp.LpVariable.dicts("Yield", (i for i in range(1, 6)), cat="Continuous", lowBound=0)

# Define objective function
prob_two_weeks += pulp.lpSum(x_two_weeks[i, j, w] for i in range(1, 6) for j in range(1, 6) for w in range(2)), "Total_Production"

# Define initial setup constraints
for i in range(1, 6):
    prob_two_weeks += x_two_weeks[i, initial_setup[i], 0] == initial_demand[i - 1], f"Initial_Setup_{i}"

# Define demand constraints for both weeks
for w in range(2):
    for i in range(1, 6):
        prob_two_weeks += pulp.lpSum(x_two_weeks[i, j, w] for j in range(1, 6)) >= (initial_demand if w == 0 else second_week_demand)[i - 1], f"Demand_{i}_{w}"

# Define machine capacity constraints for both weeks
for w in range(2):
    for i in range(1, 6):
        prob_two_weeks += pulp.lpSum(x_two_weeks[j, i, w] for j in range(1, 6)) <= machine_capacity[i - 1], f"Machine_Capacity_{i}_{w}"


prob_two_weeks.solve()

print("Status:", pulp.LpStatus[prob_two_weeks.status])
print("\nOptimal Production Schedule for Two Weeks:")
for i in range(1, 6):
    for j in range(1, 6):
        for w in range(2):
            print(f"Machine {j}, Part {i}, Week {w + 1}: {pulp.value(x_two_weeks[i, j, w])}")

print("\nTotal Production Quantity for Two Weeks:", pulp.value(prob_two_weeks.objective))
```
